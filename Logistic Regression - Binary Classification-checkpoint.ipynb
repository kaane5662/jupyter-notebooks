```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris



df = pd.read_csv("Healthcare-Diabetes.csv")
df.dropna()

plt.scatter(df["Glucose"], df["Outcome"], marker = "+")
print(sns.heatmap(df.corr()))

# iris = load_iris()
# df2 = pd.DataFrame(iris.data, columns = iris.feature_names)
# df2.head()
# print(iris.target)
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

x = np.array(df[["Glucose"]]).reshape(-1,1)
y = df["Outcome"]

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = .2, random_state=0)
model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

```

```python
from sklearn.metrics import classification_report, accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
class_report = classification_report(y_test, y_pred)
print(class_report)
```

```python
def predict(arr):
    input_data = np.array(arr).reshape(-1,1)
    predictions = model.predict(input_data)
    print(predictions)
    
predict(np.array([50,125,100,25, 150, 200,350,400,100, 135, 140, 145]))
```
